# 07 Strategy Research

Purpose: show a research-template approach for systematic strategies without profitability promises.

Data source: available processed datasets. A simple strategy template runs only if real price data is available.


## Methodology Summary

Strategies in this project are templates for hypothesis testing. They should be evaluated with costs, turnover, capacity, train/test splits, walk-forward validation and failure analysis.


```python
from __future__ import annotations

import json
from pathlib import Path
import sys

import pandas as pd
import plotly.express as px

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT / 'src'))

from russian_markets_lab.data.io import read_processed_dataset, read_processed_metadata

pd.set_option('display.max_columns', 80)
pd.set_option('display.width', 140)

def load_dataset(name: str) -> tuple[pd.DataFrame, dict]:
    df = read_processed_dataset(name)
    metadata = read_processed_metadata(name)
    print(f'{name}: {len(df)} rows, {len(df.columns)} columns')
    if metadata:
        print('generated_at:', metadata.get('generated_at'))
        print('source:', metadata.get('source'))
    else:
        print('metadata: missing')
    return df, metadata

def show_missing(name: str) -> None:
    print(f'No processed data found for {name}. Run the corresponding CLI pipeline first.')

from russian_markets_lab.strategies.momentum import generate_signals as momentum_signals
```

```python
universe, universe_metadata = load_dataset('market_universe')
liquidity, liquidity_metadata = load_dataset('liquidity_radar')
if universe.empty and liquidity.empty:
    print('No processed research data available. Build datasets before running strategy research.')
else:
    display(universe.head(10) if not universe.empty else liquidity.head(10))
```

```python
price_columns = [c for c in ['begin', 'close', 'secid'] if c in universe.columns]
if {'close'}.issubset(universe.columns):
    signals = momentum_signals(universe.rename(columns={'last_close': 'close'}))
    display(signals.head())
else:
    print('No real price panel is available in processed datasets. Strategy code is not executed to avoid fabricating backtest results.')
```

## Why This Strategy May Fail

- Edge instability.
- High turnover.
- Capacity constraints.
- Cost sensitivity.
- Liquidity decay.
- Regime change.
- Overfitting.
